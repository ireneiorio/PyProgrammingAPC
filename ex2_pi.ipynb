```python
#define function to extract one random number from uniform
def extract_coord(lower:float, higher:float):
    """
        function to extract one random numbers from interval between lower and higher
    """
    x = random.uniform(lower, higher)
    print('extracted x', x)
    return x
```

```python
#function to extract couple of numbers from uniform
def extract_coords(lower:float, higher:float):
    """
        function to extract two random numbers from interval between lower and higher (integers)
        add check if not integers!
    """
    x = random.uniform(lower, higher)
    y = random.uniform(lower, higher)
    #print('extracted')
    return [x, y]
```

```python
extract_coords(0., 1.)
```

```python
def extract_full_coords(lower:float, higher:float, nsteps:int)->tuple:
    #output tuple: list of coordinates
    coords = [] #empty list
    for i in range(0, nsteps, 1):
        coords.append(extract_coords(lower, higher))
    print('exited for loop')
    print('lenght of list is:%i'%len(coords))
    return coords  
```
